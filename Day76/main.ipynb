```python
import pandas as pd
import plotly.express as px
```

```python
df_apps = pd.read_csv('apps.csv')
```

```python
columns = df_apps.shape[1]
rows = df_apps.shape[0]
f'It has {columns} columns and {rows} rows.'
```

```python
df_apps.drop(columns=['Last_Updated', 'Android_Ver'], inplace=True)
df_apps
```

```python
df_apps_clean = df_apps.dropna()
df_apps_clean
```

```python
df_apps_clean.duplicated().value_counts()
```

```python
df_apps_clean = df_apps_clean.drop_duplicates(subset=['App', 'Type', 'Price'])
df_apps_clean.duplicated().value_counts()
```

```python
ratings = df_apps['Content_Rating'].value_counts()
ratings
```

```python
fig = px.pie(labels=ratings.index, values=ratings.values, title='Content Rating', names=ratings.index, hole=0.6)
fig.update_traces(textposition='inside', textfont_size=15, textinfo='percent')
fig.show()
```

```python
df_apps_clean['Installs'] = df_apps_clean['Installs'].astype(str).str.replace(',', '')
df_apps_clean['Installs'] = pd.to_numeric(df_apps_clean['Installs'])
df_apps_clean[['App', 'Installs']].groupby('Installs').count()
```

```python
df_apps_clean['Price'].describe()
```

```python
df_apps_clean['Price'] = df_apps_clean['Price'].astype(str).str.replace('$', '')
df_apps_clean['Price'] = pd.to_numeric(df_apps_clean['Price'])
```

```python
df_apps_clean.sort_values('Price', ascending=False).head(20)
```

```python
df_apps_clean = df_apps_clean[df_apps_clean['Price']<250]
```

```python
df_apps_clean.sort_values('Price', ascending=False)
```

```python
df_apps_clean['Revenue_Estimate'] = df_apps_clean['Price'] * df_apps_clean['Installs']
```

```python
free_vs_paid = df_apps_clean.groupby(["Category", "Type"], as_index=False).agg({'App': pd.Series.count})
df_free_vs_paid = pd.DataFrame(free_vs_paid)
```

```python
df_free_vs_paid
```

```python
bar = px.bar(df_free_vs_paid, x='Category', y='App', title='Free vs Paid Apps by Category', color='Type', barmode='group')
bar.update_layout(xaxis_title='Category', yaxis_title='Number of Apps', xaxis={'categoryorder':'total descending'}, yaxis=dict(type='log'))
bar.show()
```
